```python
from openpyxl import Workbook, load_workbook
from pathlib import Path
from datetime import datetime, timedelta
import os

import win32com.client

def salvarAnexo(message, output_dir):
    nome_arquivo = ""
    if(message.body.find("Gerência de Clientes") != -1):
        posicao_nome = message.body.find("Gerência de Clientes")
        nome_arquivo = "Base de Clientes" + message.body[posicao_nome+39] + message.body[posicao_nome+40]

        #Iterando sobre os anexos
        attachments = message.Attachments
        for attachment in attachments:
            attachment.SaveAsFile(output_dir / (nome_arquivo + ".xlsx"))

        wb = load_workbook(output_dir /(nome_arquivo + ".xlsx"))

        ws = wb.active
        #A1:F3 G1:L3 M1:Q3
        ws.unmerge_cells(start_row = 1, start_column = 1, end_row = 3, end_column = 6)
        ws.unmerge_cells(start_row = 1, start_column = 7, end_row = 3, end_column = 12)
        ws.unmerge_cells(start_row = 1, start_column = 13, end_row = 3, end_column = 17)
        ws.unmerge_cells(start_row = 4, start_column = 1, end_row = 4, end_column = 17)

        ws.delete_rows(1,4)

        wb.save(output_dir /(nome_arquivo + ".xlsx"))

    elif(message.body.find("Escritório de Gerenciamento de Clientes") != -1):
        posicao_nome = message.body.find("Escritório de Gerenciamento de Clientes")
        nome_arquivo = "Base de Clientes ESCRITORIO"

        #Iterando sobre os anexos
        attachments = message.Attachments
        for attachment in attachments:
            attachment.SaveAsFile(output_dir / (nome_arquivo + ".xlsx"))

        wb = load_workbook(output_dir /(nome_arquivo + ".xlsx"))

        ws = wb.active
        #A1:F3 G1:L3 M1:Q3
        ws.unmerge_cells(start_row = 1, start_column = 1, end_row = 3, end_column = 6)
        ws.unmerge_cells(start_row = 1, start_column = 7, end_row = 3, end_column = 12)
        ws.unmerge_cells(start_row = 1, start_column = 13, end_row = 3, end_column = 17)
        ws.unmerge_cells(start_row = 4, start_column = 1, end_row = 4, end_column = 17)

        ws.delete_rows(1,4)

        wb.save(output_dir /(nome_arquivo + ".xlsx"))

        
def consolidarBase(nome_pasta, output_dir):
    
    #Criando a base consolidada
    bTotal = Workbook()
    sTotal = bTotal.active
    sTotal.title = "Dados"
    
    linhaComeco = 0
    aux = 1
    dest_filename = "Base de Clientes POSICAO "+ nome_pasta + ".xlsx"
    for diretorio, subpastas, arquivos in os.walk(output_dir):
        for arquivo in arquivos:
            if(arquivo != dest_filename):
                wb = load_workbook(output_dir / arquivo)
                ws = wb.worksheets[0]
                linhas = ws.max_row
                colunas = ws.max_column
                if(aux == 1):
                    for i in range(1, linhas+1):
                        for j in range(1, colunas+1):
                            c = ws.cell(row = i, column = j)
                            sTotal.cell(row = i, column = j).value = c.value
                else:
                    for i in range(1, linhas+1):
                        for j in range(1, colunas+1):
                            if(i != 1):
                                c = ws.cell(row = i, column = j)
                                sTotal.cell(row = i+linhaComeco-aux+1, column = j).value = c.value
                linhaComeco += linhas    
                aux += 1
    
    bTotal.save(output_dir / dest_filename)
    

#Obtenção da data para o nome da pasta
dia = datetime.now().day
if(dia < 10): 
    dia = "0" + str(dia)
else: 
    dia = str(dia)

mes = datetime.now().month
if(mes < 10):
    mes = "0" + str(mes)
else: 
    mes = str(mes)
    
ano = str(datetime.now().year)

nome_pasta = dia + "." + mes + "." + ano

#Criação da pasta que irá receber os arquivos
output_dir = Path.cwd() / nome_pasta
output_dir.mkdir(parents=True, exist_ok=True)

#Conexão com o Outlook
outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")

#Especificando a pasta que irá ser acessada
folder = outlook.GetDefaultFolder(6).Folders("Base de Clientes")


#Pegando as mensagens
messages = folder.Items

#Limitar a emails recebidos nas ultimas 24 horas
date = datetime.now() - timedelta(days=1)
date = date.strftime('%m/%d/%Y %H:%M %p')
messages = messages.Restrict("[ReceivedTime] >= '" + date + "'")

#Iterando sobre as mensagens
for message in messages:
    salvarAnexo(message, output_dir)
    
consolidarBase(nome_pasta, output_dir)
```
